```r
library(ggplot2) 
library(dplyr)
library(stringr)
```

```r
df <- read.csv2("https://raw.githubusercontent.com/RLadiesMedellin/Meetup2-Tallerggplot2/master/Data/EncuestaJuventud.csv", 
                header = TRUE, sep = ";", dec = ",")

# Renombrar algunas columnas
df <- df %>% rename(Genero = "Sexo", 
                    ParticipacionActividadesComunitarias = "ActividadesComunitarias")

# Poner todos los nombres de columnas en minusculas
colnames(df) <- str_to_lower(colnames(df))

head(df)
```

## Grafiquemos el gasto semanal en rumba de los jóvenes según su edad y su género

### Capa 1: Data

Esta capa especifica los datos que se desean graficar

Captura de pantalla 2022-05-11 075554.png

```r
ggplot(data = df)
```

### Capa 2: Aesthetics

Esta capa especifica cómo queremos mapear nuestros datos en las escalas de la gráfica

Captura de pantalla 2022-05-11 075912.png

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba))
```

### Capa 3: Geometries

Esta capa especifica el tipo de gráfico que queremos construir

Captura de pantalla 2022-05-11 080221.png

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point()
```

### Capa 4: Layers

Esta capa permite crear sub-gráficos dentro del mismo objeto gráfico


Captura de pantalla 2022-05-11 080605.png

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero)
```

### Capa 5: Statistics

Esta capa permite graficar valores estadísticos calculados a partir de los datos

Captura de pantalla 2022-05-11 080826.png

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") 
```

### Capa 6: Coordinates

Esta capa permite ajustar las coordenadas x, y

Captura de pantalla 2022-05-11 081128.png

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
coord_cartesian(xlim = c(12, 28), ylim = c(0, 150000))
```

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
scale_y_continuous("Gasto semanal en Rumba", breaks = seq(0, 125000, 10000))
```

### Capa 7: Theme

Esta capa permite modificar todos los elementos de la gráfica diferentes a los datos (ambiar las etiquetas de los ejes x o y, agregar un título de gráfico, modificar un título de leyenda, agregar texto en cualquier parte del gráfico, cambiar el color de fondo, las líneas del eje, las líneas del gráfico)

Captura de pantalla 2022-05-11 081513.png

Hay tres tipos de elementos dentro de la capa de temas; texto, línea y rectángulo




*   Cualquier elemento de texto se puede modificar con `element_text()`
*   Cualquier elemento de línea se puede modificar con `element_line()`
*   Cualquier elemento del rectángulo se puede modificar con `element_rect()`












```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
scale_y_continuous("Gasto semanal en Rumba", breaks = seq(0, 125000, 10000)) +
theme(axis.text = element_text(color = "red", size = 14))
```

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
scale_y_continuous("Gasto semanal en Rumba", breaks = seq(0, 125000, 10000)) +
theme(axis.text.x = element_text(color = "red", size = 14))
```

```r
theme_sara <- theme(axis.text.x = element_text(color = "red", size = 14),
                    panel.grid = element_blank(),
                    panel.background = element_rect(fill = "pink"))

ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
scale_y_continuous("Gasto semanal en Rumba", breaks = seq(0, 125000, 10000)) +
theme_sara
```

#### Temas incorporados en ggplot2

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
scale_y_continuous("Gasto semanal en Rumba", breaks = seq(0, 125000, 10000)) +
theme_bw()
```

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) + 
geom_point() + 
facet_wrap(~ genero) +
stat_smooth(method = "lm", se = FALSE, col = "red") +
scale_y_continuous("Gasto semanal en Rumba", breaks = seq(0, 125000, 10000)) +
theme_linedraw()
```

### Añadir más variables en la capa Aesthetics

Podemos personalizar más nuestro gráfico añadiendo más variables en los argumentos estéticos de color o tamaño en la capa Aesthetics

```r
ggplot(data = df, aes(x = edad, y = gastosemanalrumba)) +
       geom_point()
```

```r
# Añadir variable genero por color
ggplot(data = df, aes(x = edad, y = gastosemanalrumba, color = genero)) +
       geom_point()
```

```r
# Añadir variable gastosemanaltransporte por tamaño
ggplot(data = df, aes(x = edad, y = gastosemanalrumba, size = gastosemanaltransporte)) +
       geom_point()
```
